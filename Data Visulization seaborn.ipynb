# which company have highest sales in thousands

```python
# which company have highest sales of vechicle
a=neww.sort_values(by='Sales_in_thousands' ,ascending=False)
b=sns.barplot(x="Manufacturer", y="Sales_in_thousands" ,data=a.head(12),errorbar=None)
for i in b.containers:
    b.bar_label(i,)
plt.xticks(rotation = 45, ha = 'right')
# to remove this info of botom us 
plt.plot()
```

# compare Fuel_capacity with Fuel_efficiency of vehicle

```python
 # compare Fuel_capacity with Fuel_efficiency
b=sns.barplot(x="Fuel_capacity", y="Fuel_efficiency" ,data=neww.head(50),errorbar=None)
for i in b.containers:
    b.bar_label(i,)
plt.xticks(rotation = 60, ha = 'right')
plt.show()
```

#  according to engine size , which size of engine have highest sales

```python
top_en=neww.sort_values(by='Engine_size',ascending=False)
b=sns.barplot(x="Engine_size", y="Sales_in_thousands",data=top_en.head(10),errorbar=None)
for i in b.containers:
    b.bar_label(i,)
plt.xticks(rotation = 90, ha = 'right')
```

# which type of vehicle have most sales using barplot

```python

a=sns.barplot(x="Vehicle_type", y="Sales_in_thousands" ,data=neww,errorbar=None)
for i in a.containers:
    a.bar_label(i,)
```

# which company have made the vehicle with most horse power

```python
top_hp=neww.sort_values(by='Horsepower',ascending=False)
top_hp
b=sns.barplot(x='Manufacturer',y='Horsepower',data=top_hp.head(6))
for i in b.containers:
    b.bar_label(i,)
```

# companies with lowest sales

```python

a=neww.groupby('Manufacturer')
b=a.sum().sort_values(by='Sales_in_thousands' ,ascending=True)
b
```

```python

top_five=b.head()
top_five
a=sns.barplot(x=top_five.index ,y='Sales_in_thousands', data=top_five)
for i in a.containers:
    a.bar_label(i,)
```

# which company have highest resale value ?

```python

top_resale=neww.sort_values(by='__year_resale_value',ascending=False)
a=sns.barplot(x='Manufacturer',y='__year_resale_value',data=top_resale.head(10),hue='Vehicle_type',errorbar=None)
for i in a.containers:
    a.bar_label(i,)
plt.xticks(rotation = 55, ha = 'right')
plt.plot()
```
